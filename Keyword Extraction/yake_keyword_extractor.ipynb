```python
import yake
from PyPDF2 import PdfReader
import pandas as pd
```

```python
with open('pdfs\murakami.pdf', 'rb') as pdf_file:
    # Crie um objeto PDFReader
    pdf_reader = PdfReader(pdf_file)

    # Inicialize uma string vazia para armazenar o texto extraído
    extracted_text = ""

    # Itere pelas páginas do PDF e extraia o texto
    for page_num in range(len(pdf_reader.pages)):
        page = pdf_reader.pages[page_num]
        extracted_text += page.extract_text()

# Imprima o texto extraído
print(extracted_text)
```

```python
kw_extractor = yake.KeywordExtractor()
language = "pt"
max_ngram_size = 4
deduplication_threshold = 0.9
deduplication_algo = 'seqm'
numOfKeywords = 30
windowSize = 1
features = None

custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_threshold, dedupFunc=deduplication_algo, top=numOfKeywords, features=features, windowsSize=windowSize)
```

```python
keywords = custom_kw_extractor.extract_keywords(extracted_text)
```

```python
for kw in keywords:
    print(kw)
```

```python
df = pd.DataFrame(columns=['title', 'keywords'])
for kw in keywords:
    df = pd.concat([df, pd.DataFrame([{'title': kw[0], 'keywords': kw[1]}])], ignore_index=True)
```

```python
# Export pd as CSV

df.to_csv('yake_keywords.csv', index=False, header=True, sep='\t')
```
